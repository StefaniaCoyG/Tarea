```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

###Punto 1:###

```python
def fibonacci(n):
    a, b = 0, 1 # A va a tener el valor de 0 y B el valor de 1 ya que la secuencia Fibonacci empieza con esos 2 numeros
    fib_sequence = [] #esto de aquí es una lista vacía para que vaya guardando los datos y pueda graficar correctamente

    for i in range(n):
        fib_sequence.append(a)
        c = a + b
        a = b
        b = c

    return fib_sequence

# Generar los primeros 20 números de Fibonacci
n = 20
fib_numbers = fibonacci(n)

# Graficar la sucesión
plt.plot(range(n), fib_numbers, marker="o", linestyle="-", color="b", label="Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Sucesión de Fibonacci")
plt.legend()
plt.grid()
plt.show()
```

###Punto 2:###

```python
def analyze_temperatures(temperatures):
    mean_temp = np.mean(temperatures)
    median_temp = np.median(temperatures)
    std_dev_temp = np.std(temperatures)
    print(f"Media: {mean_temp:.2f}, Mediana: {median_temp:.2f}, Desviación estándar: {std_dev_temp:.2f}")

    months = np.arange(1, len(temperatures) + 1)
    plt.plot(months, temperatures, marker='o', linestyle='-', color='r')
    plt.title('Temperatura Media Mensual')
    plt.xlabel('Meses')
    plt.ylabel('Temperatura (°C)')
    plt.grid()
    plt.show()


# Ejecución de las funciones
temperatures = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

analyze_temperatures(temperatures)
```

```python
# Definir los puntos
x1, y1 = 2, 3  # Punto 1 (x1, y1)
x2, y2 = 6, 7  # Punto 2 (x2, y2)

# Calcular la pendiente
m = (y2 - y1) / (x2 - x1)  # Fórmula de la pendiente

# Generar valores de x para la recta
x = np.linspace(x1, x2, 100)  # 100 valores entre x1 y x2
y = m * (x - x1) + y1  # Ecuación de la recta

# Graficar los puntos y la recta
plt.plot(x, y, label="Recta", color="blue")  # Dibuja la recta
plt.scatter([x1, x2], [y1, y2], color="red", label="Puntos")  # Dibuja los puntos

# Configurar el gráfico
plt.title("Gráfica de una recta entre dos puntos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()
```

```python
#Ecuaciones:
#Velocidades iniciales: vºx = vºcos(0) y vºx = vºsen(0)
#Posiciones en funcion del tiempo: x = vºcos(0)t y y = vºsin(0)t-(1/2)gt**2
#Tiempo de vuelo: t_vuelo = (2vºsin(0))/g
def trayectoria_proyectil(v0, angulos, g=9.81):
    t_max = (2 * v0 * np.sin(np.radians(max(angulos)))) / g
    t = np.linspace(0, t_max, num=100)

    plt.figure()
    for angulo in angulos:
        theta = np.radians(angulo)
        x = v0 * np.cos(theta) * t
        y = v0 * np.sin(theta) * t - 0.5 * g * t**2
        plt.plot(x, y, label=f'{angulo}°')

    plt.xlabel('Distancia (m)')
    plt.ylabel('Altura (m)')
    plt.title('Trayectoria de un proyectil')
    plt.legend()
    plt.grid()
    plt.show()

# Parámetros iniciales
velocidad_inicial = 20  # m/s
angulos_lanzamiento = [30, 45, 60]  # grados

trayectoria_proyectil(velocidad_inicial, angulos_lanzamiento)
```

```python
from numpy.random import random as rng

n = 10000 #número de interacciones

#arreglos donde se guardamos las coordenas de los puntos medios
coordenadasx = []
coordenadasy = []

#El punto donde comienza el codigo
seedx = 0.5
seedy = 0.65

# Define puntosx and puntosy
puntosx = []
puntosy = []

#Primer punto en el arreglo
puntosx.append(seedx)
puntosy.append(seedy)

#variables para mas adelante
mediox = 0
medioy = 0

#coordenadas de los vertices
v1x = 0
v2x = 0.5
v1y = 0
v2y = np.sqrt(3)/2
v3x = 1
v3y = 0

r = []
r = rng(n)

for i in range(n):
  #Elegir verices alazar
  if (r[i] <1/3):
    verticex = v1x
    verticey = v1y
  elif (1/3<r[i]<2/3):
    verticex = v2x
    verticey = v2y
  elif (r[i]>2/3):
    verticex = v3x
    verticey = v3y

  #primera interaccion con seed (punto incial)
  if (i==0):
    mediox = (verticex+mediox)/2
    medioy = (verticey+medioy)/2
  else: #Puntos medios generados por ciclos
    mediox = (verticex+mediox)/2
    medioy = (verticey+medioy)/2
  #Coordenadas de todos los puntos medios (guardar)
  puntosx.append(mediox)
  puntosy.append(medioy)

#Graficacion:
plt.title('Triángulo de Sierpinsky')
plt.plot(puntosx, puntosy, 'm ,')
plt.text(0.1, 0.7, r'$n=10000$')
plt.show()
```
